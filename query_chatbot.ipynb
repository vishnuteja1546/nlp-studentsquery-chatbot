```python
import json
import random
import nltk
import gradio as gr
from nltk.tokenize import word_tokenize

nltk.download('punkt', quiet=True)

with open('intents.json', 'r') as file:
    data = json.load(file)

def preprocess_text(text):
    """Tokenizes text and converts it to lowercase."""
    tokens = word_tokenize(text.lower())
    return tokens

def get_response(user_input):
    """Finds the best matching intent based on word overlap."""
    user_words = preprocess_text(user_input)

    best_tag = None
    max_overlap = 0

    for intent in data['intents']:
        for pattern in intent['patterns']:
            pattern_words = preprocess_text(pattern)
            overlap = len(set(user_words) & set(pattern_words))

            if overlap > max_overlap:
                max_overlap = overlap
                best_tag = intent['tag']

    if best_tag and max_overlap > 0:
        for intent in data['intents']:
            if intent['tag'] == best_tag:
                return random.choice(intent['responses'])

    return "I'm sorry, I didn't quite catch that. Could you rephrase?"


def chatbot_ui(message, history):
    return get_response(message)


demo = gr.ChatInterface(
    fn=chatbot_ui,
    title="Customer Service Assistant",
    description="Ask me about our opening hours, location, or say hello!",
    examples=["Hi", "What are your hours?", "Can you help me?"]
)

demo.launch(share=True, debug=True)
```
